Clustering

1. Audio Feel Clustering using Kmeans

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Columns for clustering
cols = ["danceability", "energy", "valence"]
data = df[cols].fillna(df[cols].median())

# Scale
scaler = StandardScaler()
scaled = scaler.fit_transform(data)

# K-Means
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df["cluster_audio_feel"] = kmeans.fit_predict(scaled)

# Cluster counts
cluster_counts = df["cluster_audio_feel"].value_counts().sort_index()

# Custom colors
colors = ["#FF6F61", "#6B5B95", "#88B09B", "#F7CAC9"]

# Line chart
plt.plot(cluster_counts.index, cluster_counts.values, marker='o', color="#6B5B95", linewidth=2, markersize=8)
for i, count in enumerate(cluster_counts.values):
    plt.scatter(cluster_counts.index[i], count, color=colors[i], s=100, label=f"Cluster {cluster_counts.index[i]}")
plt.xlabel("Cluster")
plt.ylabel("Number of Songs")
plt.title("Audio Feel Clusters (Line Chart)", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("spotify dataset.csv")

# Select features for "audio feel"
audio_features = ['danceability', 'energy', 'valence', 'tempo', 'loudness', 'acousticness', 'liveness', 'speechiness']
X = df[audio_features].fillna(df[audio_features].mean())

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal clusters using elbow method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Audio Feel Clustering')
plt.show()

# Fit KMeans with chosen clusters (e.g., k=4)
kmeans = KMeans(n_clusters=4, random_state=42)
df['AudioFeelCluster'] = kmeans.fit_predict(X_scaled)

# Show sample results
print(df[['AudioFeelCluster'] + audio_features].head())
```

2. Vocality / Vocal Clustering

```python
cols = ["speechiness", "acousticness", "loudness"]
data = df[cols].fillna(df[cols].median())

scaled = StandardScaler().fit_transform(data)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["cluster_vocality"] = kmeans.fit_predict(scaled)

pca = PCA(n_components=2)
reduced = pca.fit_transform(scaled)

plt.scatter(reduced[:,0], reduced[:,1], c=df["cluster_vocality"], cmap="plasma", s=15)
plt.title("Vocality Clusters")
plt.show()
```

3. Rhythm Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Columns for clustering
cols = ["tempo", "liveness", "instrumentalness"]
data = df[cols].fillna(df[cols].median())

# Scale
scaled = StandardScaler().fit_transform(data)

# K-Means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df["cluster_rhythm"] = kmeans.fit_predict(scaled)

# Cluster counts
cluster_counts = df["cluster_rhythm"].value_counts()

# Custom colors
colors = ["#FF6F61", "#6B5B95", "#88B04B", "#F7CAC9", "#92A8D1"]

# Pie chart
plt.pie(
    cluster_counts,
    labels=[f"Cluster {i}" for i in cluster_counts.index],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)
plt.title("Rhythm Clusters Distribution", fontsize=14)
plt.show()

# Bar chart
cluster_counts.plot(
    kind='bar',
    color=colors,
    edgecolor='black'
)
plt.xlabel("Cluster")
plt.ylabel("Number of Songs")
plt.title("Rhythm Clusters Count", fontsize=14)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select features
X = df[['danceability', 'energy']].fillna(df[['danceability', 'energy']].mean())

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Run KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster_audio_feel'] = kmeans.fit_predict(X_scaled)

# Count songs per cluster
cluster_counts = df['cluster_audio_feel'].value_counts()

# Pie chart
plt.pie(cluster_counts, labels=cluster_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'lightgreen'])
plt.title("K-Means Clusters (Danceability + Energy)")
plt.show()
```

Distribution and Trends

```python
df['LoudnessCategory'] = pd.cut(
    df['loudness'],
    bins=3,
    labels=['Soft', 'Moderate', 'Loud']
)
df.groupby('LoudnessCategory')['energy'].mean()
df.groupby('LoudnessCategory')['energy'].mean().plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.ylabel("Average Energy")
plt.title("Energy per Loudness Category")
plt.show()

```

```python
import pandas as pd

# Fill missing values before binning
df['tempo'] = df['tempo'].fillna(df['tempo'].median())

# Create tempo category bins
df['TempoCategory'] = pd.cut(
    df['tempo'],
    bins=[0, 90, 120, 300],  # Increased upper limit to cover higher tempos
    labels=['Slow', 'Medium', 'Fast']
)

# Check the new column
print(df['TempoCategory'].value_counts())
df['TempoCategory'].value_counts().plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.xlabel("Tempo Category")
plt.ylabel("Number of Songs")
plt.title("Songs by Tempo Category")
plt.show()
```
